### 10. Add new column as HL_Calories where more than average calories should be categorized as HIGH and less than average calories should be categorized as LOW

```python
output = []
for i in cereals.calories:
    if i > cereals.calories.mean():
        output.append('HIGH')
    else:
        output.append('LOW')
        
```

```python
cereals.calories.mean()
```

```python
cereals['HL_Calories'] = output
```
