# COVID-19 EDA Report

## Dataset: 
### The dataset is collected from the World Health Organization (WHO) website. It contains daily reported COVID-19 cases and deaths for each country and WHO region, covering the period from 4 January 2020 to 3 August 2025.

## Objective: 
### The objective of this project is to analyze the COVID-19 dataset to understand its global impact. The project aims to identify the most affected countries, the countries with the highest deaths, and the countries that remained relatively safer. It also focuses on studying the trends of cases and deaths over time and comparing the impact across different WHO regions.

```python
# Import the library

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import the dataset and check its structure

df = pd.read_excel(r"D:\Imarticus\PGDA_43\Python\EDA_Project\COVID-19 Dataset.xlsx")
```

```python
# Check the shape of the dataset

df.shape

print(f'The number of columns is : {df.shape[1]} ')
print(f'The number of rows is : {df.shape[0]} ')
```

## Data cleaning and Transformation

```python
# check the null values

df.isnull().sum()
```

### Handling Missing Values:
### In the dataset, the missing values in New_cases and New_deaths mean that no cases or deaths were reported on those days, so we will fill them with zero (0). Also, since the Country column has no missing values, the Country_code column is not important for the analysis, so we will drop it.

```python
df.fillna({'New_cases' : 0,'New_deaths' : 0}, inplace= True)
```

```python
df.drop(columns = 'Country_code', inplace= True)
```

```python
df.isnull().sum()
```

```python
# check the number of duplicates values

df.duplicated().sum()
```

```python
# check the descriptive statistics

df.describe()
```

### New_cases and New_deaths initially contain negative values. Since counts cannot be negative, we are replacing all negative entries with 0 while keeping positive values unchanged.

```python
df['New_cases'] = df['New_cases'].clip(lower= 0)
```

```python
df.New_cases.min()
```

```python
df['New_deaths'] = df['New_deaths'].clip(lower= 0)
```

```python
df.New_deaths.min()
```

```python
# Check the Data type of all Columns

df.info()
```

```python
# Convert the data type of 'Date_reported' to datetime and 'New_cases' & 'New_deaths' to int
```

```python
df['New_cases'] = df['New_cases'].astype('int')
df['New_deaths'] = df['New_deaths'].astype('int')
```

```python
df['Date_reported'] = pd.to_datetime(df['Date_reported'])
```

```python
df.info()
```

```python
# Check the Unique Value of 'Country' and 'WHO Region'

df[['Country','WHO_region']].nunique()
```

```python
# Group the data by Country, Region and Date

country = df.groupby('Country',as_index= False)[['New_cases','New_deaths']].sum()
```

```python
region = df.groupby('WHO_region',as_index= False)[['New_cases','New_deaths']].sum()
```

```python
date = df.groupby('Date_reported',as_index= False)[['New_cases','New_deaths']].sum()
```

## Total COVID-19 Cases, Deaths and Death ratio

```python
total_cases = df['New_cases'].sum()
total_deaths = df['New_deaths'].sum()
death_ratio = (total_deaths/total_cases)*100
```

```python
print(f'Total COVID-19 case records in the dataset: {total_cases:,}')
print(f'Total COVID-19 death records in the dataset: {total_deaths:,}')
print(f'COVID-19 death ratio: {death_ratio:.2f}%')
```

## Trend of COVID-19 Cases Over Time

```python
sns.set_theme(style= 'whitegrid', palette= 'winter_r', font_scale= 1.1)
```

```python
# Change to Y-Axis in Million

from matplotlib.ticker import FuncFormatter 
def millions(x,pos):
    return f'{x/1_000_000:.1f}M'
```

```python
plt.figure(figsize= (12,6))
sns.lineplot(x = 'Date_reported', y = 'New_cases', data = date)
plt.title('COVID-19 Cases Trend Worldwide', fontsize = 20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Number of Cases', fontsize = 15)
ax = plt.gca()
ax.yaxis.set_major_formatter(FuncFormatter(millions))
plt.show()
```

### Insight: 
COVID-19 cases started in mid-2020 and increased slowly. In early 2022, cases
rose quickly and crossed 8 million. After that, they decreased but again went up near the
end of 2022, reaching around 8 million before finally slowing down and coming under
control.

## Trend of COVID-19 Deaths Over Time

```python
plt.figure(figsize= (12,6))
sns.lineplot(x= 'Date_reported', y= 'New_deaths', data = date)
plt.title('COVID-19 Deaths Trend Worldwide', fontsize = 20)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Number of Deaths', fontsize=15)
plt.show()
```

### Insight: 
Daily COVID-19 deaths peaked dramatically in 2020-2022 but then fell sharply
and have remained low since 2023, showing the pandemic's transition to a manageable
endemic phase.

## WHO Regions with Death Ratio

```python
region['death_ratio'] = (region['New_deaths']/region['New_cases'])*100
```

```python
region
```

```python
plt.figure(figsize= (12,6))
sns.barplot(x= 'WHO_region', y= 'death_ratio', data= region.sort_values(by= 'death_ratio', ascending= False) )
plt.title('COVID-19 Death Ratio by WHO Region', fontsize= 20)
plt.xlabel('WHO Region',fontsize= 15)
plt.ylabel('Death Ratio in Percentage', fontsize= 15)
plt.show()
```

### Insight: 
The COVID-19 death ratio varies across WHO regions. AFR (Africa) has the
highest death ratio, followed by OTHER, AMR (Americas), and EMR (Eastern
Mediterranean). SEAR (South-East Asia) and EUR (Europe) have moderate death ratios,
while WPR (Western Pacific) has the lowest. This indicates significant regional differences
in COVID-19 mortality rates.

## Regional Contribution to Total COVID-19 Cases

```python
plt.figure(figsize= (10,6))
patches, texts, autotexts = plt.pie(region['New_cases'],labels= region['WHO_region'], autopct= '%1.1f%%', startangle= 90)
plt.setp(texts, color= 'black', fontsize = 12)
plt.setp(autotexts, color= 'black', fontsize = 12)
plt.title('COVID-19 Cases Distribution by WHO Region', fontsize = 20)
plt.axis('equal')
plt.show()
```

### Insight: 
Out of the total COVID-19 cases, the EUR region accounts for the highest share
at 36.1%, followed by WPR with 27.7% and AMR with 24.9%. The remaining regions,
SEAR, EMR, and AFR, contribute only a minor portion of the total cases. The 'Other'
category records 0% of the total cases.

## Regional Contribution to Total COVID-19 Deaths

```python
plt.figure(figsize= (10,6))
patches, texts, autotexts = plt.pie(region['New_deaths'], labels= region['WHO_region'], autopct= '%1.1f%%', startangle= 90)
plt.setp(texts, color= 'black', fontsize = 12)
plt.setp(autotexts, color= 'black', fontsize = 12)
plt.title('COVID-19 Death Distribution by WHO Region', fontsize = 20)
plt.axis('equal')
plt.show()
```

### Insight:
Out of the total COVID-19 deaths, the AMR region accounts for the highestshare at 43.4%, followed by EUR with 32.3%. The remaining regions SEAR, WPR, EMR,and AFR contribute only a minor portion of the total deaths. The 'Other' category records 0% of the total deaths.
